4. Как найти самое часто встречаемое значение между двумя массивами? 

````python
from collections import Counter

array1 = np.array([1, 2, 3, 4, 5, 4, 2, 1, 4])
array2 = np.array([4, 5, 6, 7, 4, 1, 2, 5, 7])

combined = np.concatenate((array1, array2))

counter = Counter(combined)

most_common_value = counter.most_common(1)[0][0]

print(most_common_value)
````

3. Правда ли, что в среднем девушки сдают экзамены лучше, чем мужчины

````python
from scipy import stats

male_scores = db[db['Gender'] == 'Male']['Exam_Score']
female_scores = db[db['Gender'] == 'Female']['Exam_Score']

f_stat, p_value = stats.f_oneway(male_scores, female_scores)

if p_value < 0.05:
    if female_scores.mean() > male_scores.mean():
        print("Да")
    else:
        print("Нет")
else:
    print("Различия между средними баллами мужчин и женщин не являются статистически значимыми.")
````

4. Выведите 25 и 75 перцентили времени сна студентов

5. Правда ли, что те, кто спит больше(75 перцентиль и выше) сдает экзамен лучше, чем студенты с недосыпом(25 перцентиль и меньше)

````python
low_sleep_score = db[db['Sleep_Hours'] <= sleep_25]['Exam_Score']
high_sleep_score = db[db['Sleep_Hours'] >= sleep_75]['Exam_Score']

f_stat, p_value = stats.f_oneway(low_sleep_score, high_sleep_score)

if p_value < 0.05:
    if high_sleep_score.mean() > low_sleep_score.mean():
        print("Да")
    else:
        print("Нет")
else:
    print("Различия между средними баллами многоспящих и малоспящих не являются статистически значимыми.")
````

7. Правда ли, что слуденты, чьи родители имеют высшее образование(College || Postgraduate), сдают экзамены лучше, чем остальные?

````python
high_education_group = db[db['Parental_Education_Level'].isin(['College', 'Postgraduate'])]
low_education_group = db[~db['Parental_Education_Level'].isin(['College', 'Postgraduate'])]

exam_scores_high_education = high_education_group['Exam_Score']
exam_scores_low_education = low_education_group['Exam_Score']

f_stat, p_value = stats.ttest_ind(exam_scores_low_education, exam_scores_high_education)
if p_value < 0.05:
    if exam_scores_high_education.mean() > exam_scores_low_education.mean():
        print("Да")
    else:
        print("Нет")
else:
    print("Различия между средними баллами не являются статистически значимыми.")
````

9. Правда ли, что в семьях с доходом больше среднего дети сдают экзамены лучше?

````python
above_avg_income = db[db['Family_Income'] == 'High']['Exam_Score']
below_avg_income = db[db['Family_Income'].isin(['Low', 'Medium'])]['Exam_Score']

mean_above = above_avg_income.mean()
mean_below = below_avg_income.mean()

f_stat, p_value = stats.f_oneway(above_avg_income, below_avg_income)

if p_value < 0.05:
    if mean_above > mean_below:
        print("Да")
    else:
        print("Нет")
else:
    print(print("Различия между средними баллами не являются статистически значимыми."))
````
